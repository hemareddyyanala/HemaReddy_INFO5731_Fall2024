```python
#importing the necessary libraries
import pandas as pd  # for data manipulation and analysis
from textblob import TextBlob  # for sentiment analysis
import re  # for regular expression operations
import nltk  # Natural Language Toolkit for language processing
from nltk import pos_tag, ne_chunk  # for part-of-speech tagging and named entity recognition
from nltk.tokenize import word_tokenize  # for tokenizing text into words
from nltk.corpus import stopwords  # for stop words used in text processing
# downloading stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# setting of English stopwords using the set() emthod
stop_words = set(stopwords.words('english'))  # creating a set of common English stopwords for filtering

# creating a function to count the number of words
def word_count(text):
    return len(text.split())  # splitting the text into words and return the count

# creating a function to compute the sentiment polarity of the text
def semantic_polarity(text):
    return TextBlob(text).sentiment.polarity  # use TextBlob to compute the polarity

# creating a function to classify sentiment based on polarity
def classify_sentiment(polarity):
    if polarity > 0: #if polarity is greater than zero then it is positive
        return 'positive'
    elif polarity < 0:
        return 'negative' #else if polarity is less than zero then it is negative
    else:
        return 'neutral'  #neutral if polarity is zero

#function for computing sentiment subjectivity
def sentiment_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity  # use TextBlob to compute the subjectivity of the text

# creating function to count stop words in the text
def stop_word_count(text):
    words = text.split()  # split the text into words using split() method
    return sum(1 for word in words if word.lower() in stop_words)  # count the words which are stopwords

# creating a function to count emojis in the text
def emoji_count(text):
    return len(re.findall(r'[^\w\s,]', text))  # using regex to find and count emojis or non-alphanumeric characters

# creating a function to count specific punctuation marks
def punctuation_count(text, punctuation):
    return text.count(punctuation)  # counting occurrences of the specified punctuation in the text

# making a function to extract part-of-speech tags from comments data
def extract_pos_tags(text):
    tokens = word_tokenize(text)  # Tokenize the text into words using word_tokenize()
    tagged = pos_tag(tokens)  # applying the part-of-speech tagging
    return dict(tagged)  # Return the tagged words as a dictionary

# defining a function to extract named entities in the text
def extract_named_entities(text):
    tokens = word_tokenize(text)  # again, tokenizing the text into words
    tagged = pos_tag(tokens)  # then, tagging the tokens
    entities = ne_chunk(tagged)  # applying named entity recognition using ne_chunk()
    return entities  # returning the identified named entities

#applying the feature extraction functions to the DataFrame comments_df
comments_df['Word_Count'] = comments_df['text'].apply(word_count)  # to count words in each comment
comments_df['Sentiment_Polarity'] = comments_df['text'].apply(semantic_polarity)  # to calculate polarity
comments_df['Sentiment_Subjectivity'] = comments_df['text'].apply(sentiment_subjectivity)  # for aalculating subjectivity
comments_df['Stop_Word_Count'] = comments_df['text'].apply(stop_word_count)  # to count stop words
comments_df['Emoji_Count'] = comments_df['text'].apply(emoji_count)  # helps count emojis
comments_df['Exclamation_Marks'] = comments_df['text'].apply(lambda x: punctuation_count(x, '!'))  # to count exclamation marks
comments_df['Question_Marks'] = comments_df['text'].apply(lambda x: punctuation_count(x, '?'))  # to count question marks
comments_df['POS_Tags'] = comments_df['text'].apply(extract_pos_tags)  # extracting POS tags
comments_df['Named_Entities'] = comments_df['text'].apply(extract_named_entities)  # extracting named entities
comments_df['Overall opinion'] = comments_df['Sentiment_Polarity'].apply(classify_sentiment)  # classifying the overall sentiment
```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder  # importing the MinMaxScaler and LabelEncoder from sklearn
from sklearn.feature_selection import chi2  # importing chi2 for feature selection

# First, we define the feature matrix (X)
# For this, I selected numerical feature columns from the comments DataFrame for the feature selection process
X = comments_df[['Word_Count', 'Sentiment_Polarity', 'Sentiment_Subjectivity',
        'Stop_Word_Count', 'Emoji_Count', 'Exclamation_Marks', 'Question_Marks']]

#converting the target variable 'Overall opinion' to numerical labels using LabelEncoder
label_encoder = LabelEncoder()  # Initialising the LabelEncoder
y = label_encoder.fit_transform(comments_df['Overall opinion'])  # Transforming target variable to numeric

#initialising the MinMaxScaler to scale features to a range between 0 and 1 so they are on the same scale
scaler = MinMaxScaler()
# fit_transform() adjusts the above scaler to the data and transforms the feature matrix at the same time
X_scaled = scaler.fit_transform(X)  # Apply scaling to feature matrix

#Performing Chi-Square test for feature selection using chi2(). It outputs two arrays: chi2_scores (feature importance) and p_values (significance level)
chi2_scores, p_values = chi2(X_scaled, y)
# creating a DataFrame to hold the feature names, chi2_scores, and p_values for viewing and analysis
chi2_results = pd.DataFrame({
    'Feature': X.columns,       # storing column names of the features from the comments DataFrame
    'Chi2_Score': chi2_scores,  # Chi-Square scores from the test
    'p_value': p_values         # p-values from the test
})

#sorting the DataFrame by Chi-Square scores in descending to prioritize features based on importance
chi2_results_sorted = chi2_results.sort_values(by='Chi2_Score', ascending=False)

#printing the sorted DataFrame to show which features have the highest Chi-Square scores and are the most important
print(chi2_results_sorted)
```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
import torch  #importing PyTorch for handling tensor operations
from transformers import BertTokenizer, BertModel  #importing BERT tokenizer and model from Hugging Face transformers
from sklearn.metrics.pairwise import cosine_similarity  #importing cosine similarity function to measure similarity

#loading the BERT tokenizer (here 'bert-base-uncased' is the specific model version)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
#loading the BERT model, it is pre-trained on general English text
model = BertModel.from_pretrained('bert-base-uncased')

# creating a function to get BERT embeddings for given text input
def get_bert_embedding(text):
    #tokenizing the input text which means convert text into tokens that BERT understands
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    # getting the BERT embeddings without computing gradients
    with torch.no_grad():
        outputs = model(**inputs)  #passing the tokenized inputs to the BERT model
        embeddings = outputs.last_hidden_state.mean(dim=1) # getting the mean of the last hidden layer's embeddings across all tokens using the mean() method
    return embeddings  #returning the calculated embeddings for the input text

text_embeddings = []  #creating an empty list to store BERT embeddings for each text

# using for loop to iterate through each text in the 'comments_df' DataFrame
for text in comments_df['text']:
    text_embeddings.append(get_bert_embedding(text)) #appending the embedding for each text to the list

text_embeddings = torch.cat(text_embeddings, dim=0) #converting the list of tensors (embeddings) into a single tensor (matrix of embeddings)
query = "The best Taylor Swift song" #defing an example query for which we want to find the most similar comments in the dataset
query_embedding = get_bert_embedding(query) #getting the BERT embedding for the query

#calculating cosine similarity between the query and each text embedding, it takes two matrices and computes the cosine similarity
cosine_similarities = cosine_similarity(query_embedding, text_embeddings)
#ranking the documents based on similarity (in descending order) by storing the similarity score in the 'comments_df' in a new 'Similarity' column
comments_df['Similarity'] = cosine_similarities[0]
#sorting the DataFrame by the 'Similarity' column to rank the texts in descending order
comments_df_sorted = comments_df.sort_values(by='Similarity', ascending=False)

#printing only the 'text' and 'Similarity' columns for a clearer view
print(comments_df_sorted[['text', 'Similarity']].head(10))
```

```python
#saving the ranked comments by exporting the sorted DataFrame to a CSV file named 'ranked_youtube_comments.csv'
comments_df_sorted.to_csv('ranked_youtube_comments.csv', index=False)
#printing a message that the CSV file has been saved
print("The Ranked comments have been saved to 'ranked_youtube_comments.csv'")
```
